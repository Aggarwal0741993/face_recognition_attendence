```python
import os
import pickle
import numpy as np
from insightface.app import FaceAnalysis
from PIL import Image

# Initialize InsightFace
app = FaceAnalysis(name="buffalo_l")
app.prepare(ctx_id=0, det_size=(640, 640))

# Set base directory where each subfolder is a person's name
base_path = "/content/registered_faces/FACES_PROJECT_ML"

known_faces = []

# Loop through each person's folder
for person_folder in os.listdir(base_path):
    person_path = os.path.join(base_path, person_folder)
    if not os.path.isdir(person_path):
        continue

    print(f"Processing: {person_folder}")
    image_files = [img for img in os.listdir(person_path) if img.lower().endswith(('.jpg', '.jpeg', '.png'))]
    image_files = image_files[:30]  # Limit to 20 images per person

    count = 0
    for img_file in image_files:
        img_path = os.path.join(person_path, img_file)
        img = np.array(Image.open(img_path).convert("RGB"))

        faces = app.get(img)
        if faces:
            known_faces.append({
                "name": person_folder,
                "embedding": faces[0].embedding.tolist()
            })
            count += 1
            print(f"  Saved: {img_file}")
        else:
            print(f"  No face found in {img_file}")

    print(f"Completed: {person_folder} ({count} faces saved)")

# Save all embeddings to a pickle file
with open("known_faces.pkl", "wb") as f:
    pickle.dump(known_faces, f)

print("Done: known_faces.pkl created.")

```

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js  # Add this import as it's used by eval_js
import io
import numpy as np # Import numpy
from PIL import Image # Import Image from PIL
import base64 # Import the base64 module

def capture_image():
    display(Javascript('''
        async function capture() {
            const div = document.createElement('div');
            const capture = document.createElement('button');
            capture.textContent = 'Capture';
            div.appendChild(capture);
            document.body.appendChild(div);
            const stream = await navigator.mediaDevices.getUserMedia({video: true});
            const video = document.createElement('video');
            video.srcObject = stream;
            video.play();
            div.appendChild(video);
            await new Promise((resolve) => capture.onclick = resolve);
            const canvas = document.createElement('canvas');
            canvas.width = video.videoWidth;
            canvas.height = video.videoHeight;
            canvas.getContext('2d').drawImage(video, 0, 0);
            stream.getTracks().forEach(track => track.stop());
            div.remove();
            return canvas.toDataURL();
        }
        google.colab.kernel.invokeFunction('notebook.run', [], { 'code': capture.toString() })'''))

    data_url = eval_js("capture()")
    # Use base64.b64decode instead of b64decode
    content = base64.b64decode(data_url.split(',')[1])
    return np.array(Image.open(io.BytesIO(content)))
```

```python
from datetime import datetime
import os
import pandas as pd
import pytz

def mark_attendance(name):
    ist = pytz.timezone('Asia/Kolkata')
    now = datetime.now(ist)
    today = now.strftime("%d-%m-%y")
    time_now = now.strftime("%H:%M:%S")
    file = "attendance_log.csv"

    if not os.path.exists(file):
        pd.DataFrame(columns=["Name", "Date", "Entry Time", "Exit Time"]).to_csv(file, index=False)

    df = pd.read_csv(file)

    # Filter today's records for the person
    user_today = df[(df["Name"] == name) & (df["Date"] == today)]

    if user_today.empty:
        # No entry yet → mark Entry
        df = pd.concat([df, pd.DataFrame([{
            "Name": name, "Date": today, "Entry Time": time_now, "Exit Time": ""
        }])], ignore_index=True)
        print(f"{name} — Entry marked at {time_now} (IST)")

    elif pd.isna(user_today.iloc[-1]["Exit Time"]) or user_today.iloc[-1]["Exit Time"] == "":
        # Entry exists, Exit is not marked → mark Exit
        idx = user_today.index[-1]
        df.at[idx, "Exit Time"] = time_now
        print(f"{name} — Exit marked at {time_now} (IST)")

    else:
        # Both entry and exit already marked
        print(f"{name} — Entry and Exit already marked for today.")

    df.to_csv(file, index=False)

```

```python
import cv2
import numpy as np

image = capture_image()
if image.shape[2] == 4:
    image = cv2.cvtColor(image, cv2.COLOR_RGBA2RGB)

faces = app.get(image)

for face in faces:
    emb = face.embedding
    name = "Unknown"
    max_sim = 0.6  # Adjust similarity threshold for cosine similarity

    for known_face_data in known_faces:
        # Indent the lines inside the inner loop
        known_face = known_face_data["name"]
        known_emb = np.array(known_face_data["embedding"]) # Convert list embedding to numpy array

        # Calculate similarity between the captured face embedding and the known face embedding
        # Using cosine similarity formula: dot product of embeddings divided by product of their norms
        sim = np.dot(emb, known_emb) / (np.linalg.norm(emb) * np.linalg.norm(known_emb))

        if sim > max_sim:
            # Correct the variable name from known_name to known_face
            name = known_face
            max_sim = sim

    # Now that the inner loop is complete, check the best match for the current face
    if name != "Unknown":
        print(f"Face recognized: {name} (similarity: {max_sim:.4f})")
        mark_attendance(name)
    else:
        print("Face not recognized.")
```
